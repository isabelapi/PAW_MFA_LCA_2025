```python
# Import CSV files for all flow definitions


def import_flows(base_path):
    imported_flows = []
    for flow_def in flow_definitions:
        # Create variable name: from_process_name_to_process_name
        var_name = f"{flow_def.from_process_name}_{flow_def.to_process_name}"
        imported_flows.append(var_name)
        # Create file name: from_process_name__to_process_name.csv (double underscore)
        file_name = f"{flow_def.from_process_name}__{flow_def.to_process_name}.csv"
        file_path = base_path + file_name

        # Read CSV and assign to variable
        globals()[var_name] = pd.read_csv(file_path)
    print(f"Imported flows: {imported_flows}")
```

### extract information per scenario and year

```python
# evaluate per year and scenario
years = [2020, 2035, 2045]
scenarios = ["high_env_pol", "low_env_pol"]
all_activities = pd.DataFrame()
# drop all rows whith 0 values
for flow in imported_flows:
    flow.drop(flow[flow["value"] == 0].index, inplace=True)
    # create df with year as rows and scenario as columns
    results = pd.DataFrame(columns=["flow", "scenario", "year"])

    # add data to results for flow: if column "time" does not exist, add data to results for all years
    if "time" not in flow.columns:
        for year in years:
            results.loc[year] = flow.loc[year]
    else:
        # if column "time" exists, add data to results for the specified years
        results = flow[flow["time"].isin(years)]
    all_activities = pd.concat([all_activities, results])

all_activities.head()
```
